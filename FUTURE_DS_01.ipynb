```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/Sample - Superstore.csv",encoding='latin1')
data
```

```python
data.drop_duplicates(inplace=True)
data.dropna(inplace=True)
```

```python
data
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
data['Profit Margin'] = data['Profit'] / data['Sales']
```

```python
total_revenue = data['Sales'].sum()
total_profit = data['Profit'].sum()
profit_margin = total_profit / total_revenue
total_orders = data['Order ID'].nunique()
```

```python
print(total_revenue)
print(total_profit)
print(profit_margin)
print(total_orders)
```

```python
trend = data.groupby(data['Order Date'].dt.to_period('M'))['Sales'].sum()
```

```python
category = data.groupby('Category')[['Sales','Profit']].sum()
```

```python
top_products = data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
```

```python
region = data.groupby('Region')[['Sales','Profit']].sum()
```

```python
print(trend)
print(category)
print(top_products)
print(region)
```

```python
trend = data.groupby('Month')['Sales'].sum()

plt.figure()
trend.plot()
plt.title("Revenue Trend Over Time")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
category = data.groupby('Category')[['Sales','Profit']].sum()

category.plot(kind='bar')
plt.title("Sales and Profit by Category")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
top_products = data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.xlabel("Sub-Category")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
region = data.groupby('Region')[['Sales','Profit']].sum()

region.plot(kind='bar')
plt.title("Region-wise Sales and Profit")
plt.xlabel("Region")
plt.ylabel("Value")
plt.show()
```

```python
plt.figure()
plt.scatter(data['Sales'], data['Profit'])
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs Profit")
plt.show()
```

```python
top_category = category['Sales'].idxmax()

low_profit_category = category['Profit'].idxmin()

best_region = region['Sales'].idxmax()

print("\nINSIGHTS:")
print(f"Top Revenue Category: {top_category}")
print(f"Lowest Profit Category: {low_profit_category}")
print(f"Best Performing Region: {best_region}")
```
